### üéØObjetivo dessa sec√ß√£o

Avaliar se os descontos aplicados realmente impulsionam vendas e lucro:

As principais perguntas a serem respondidas s√£o:  
‚úÖ Descontos altos realmente aumentam o volume vendido?  
‚úÖ Qual faixa de desconto gera **maior lucro**, n√£o apenas vendas?  
‚úÖ Existe desconto que **prejudica a margem**?
‚úÖ O impacto do desconto muda por pa√≠s ou segmento?   

```python
import pandas as pd
```

```python
finances = pd.read_excel('../../database/financial.xlsx')
finances.head()
```

```python
# Descontos altos realmente aumentam o volume vendido?

discount_volume = (
    finances
        .groupby("discount_band")[["units_sold", "sales", "profit"]]
        .sum()
        .reset_index()
        .sort_values("units_sold", ascending=False)
)

discount_volume
```

```python
# Qual faixa de desconto gera maior lucro, n√£o apenas vendas?

# Obter o relat√≥rio anterior para outro tipo de manipula√ß√£o
discount_performance = discount_volume.copy()

# Calculo de margem pelos descontos dados ou n√£o
discount_performance["profit_margin"] = (
    discount_performance["profit"] / discount_performance["sales"]
)

discount_performance.sort_values("profit", ascending=False)
```

```python
# Existe desconto que prejudica a margem?

discount_performance.sort_values("profit_margin")
```

```python
# O impacto do desconto muda por segmento?

columns_grouping = ['segment', 'discount_band'] 
columns_summary = ['units_sold', 'sales', 'profit'] 

discount_by_segment = (
    finances
        .groupby(by=columns_grouping)[columns_summary]
        .sum()
        .reset_index()
)

discount_by_segment["profit_margin"] = (
    discount_by_segment["profit"] / discount_by_segment["sales"]
)


sort_columns = ['segment', 'profit_margin']
discount_by_segment.sort_values(sort_columns, ascending=[True, False])

```

```python
# O impacto do desconto muda por pa√≠s?

columns_grouping = ['country', 'discount_band'] 
columns_summary = ['units_sold', 'sales', 'profit'] 

discount_by_country = (
    finances
        .groupby(by=columns_grouping)[columns_summary]
        .sum()
        .reset_index()
)

discount_by_country["profit_margin"] = (
    discount_by_country["profit"] / discount_by_country["sales"]
)

sort_columns = ['country', 'profit_margin']
discount_by_country.sort_values(sort_columns, ascending=[True, False])
```
